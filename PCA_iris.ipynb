```python
import pandas as pd 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics 

df = pd.read_csv('/home/hamza/Downloads/DataSets/iris.csv')
print(df.head())
```

```python
sns.pairplot(df)
```

```python
X = df.drop('variety', axis = 1)
y = df['variety']

ss = StandardScaler()
X = ss.fit_transform(X)

pca = PCA()
X_v = pca.fit_transform(X)
print(pca.explained_variance_ratio_)
```

```python
pca1 = PCA(n_components = 1)
X_1 = pca1.fit_transform(X)

classifier = RandomForestClassifier(max_depth = 2, random_state = 0)
classifier.fit(X_1, y)
y_pred_1 = classifier.predict(X_1)

print(metrics.confusion_matrix(y, y_pred_1))
print(metrics.accuracy_score(y, y_pred_1))
```

```python
pca2 = PCA(n_components = 2)
X_2 = pca2.fit_transform(X)

classifier.fit(X, y)
y_pred_2 = classifier.predict(X)

print(metrics.confusion_matrix(y, y_pred_2))
print(metrics.accuracy_score(y, y_pred_2))
```
